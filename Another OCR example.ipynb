```python
import pytesseract
import cv2
from PIL import Image
```

```python
img = cv2.imread("data/Screenshot (124).png") #uploading image
openImage = Image.open("data/Screenshot (124).png")
display(openImage)
```

```python
ocr_result = pytesseract.image_to_string(img)
# print(ocr_result)
ocr_result = ocr_result.split("\n")
```

```python
#manipulating initial results
results = []
starts_with = input("Enter A to Z: ")  #for searching name entities
for item in ocr_result:
    results.append(item)
    item.strip().replace("\n", " ")
    item.split("   ")[0]
    if len(item) > 2:
        if item[0] == starts_with:
            print(item)
    # if starts_with not in item:
    #     print("No results")
```
